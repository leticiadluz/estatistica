# 1 - Variáveis Aleatórias Discretas

### Variáveis aleatórias discretas são aquelas que assumem valores em um conjunto finito ou infinito enumerável de pontos. Isso significa que os valores que a variável pode assumir são distintos e podem ser contados individualmente, mas não podem ser medidos em uma escala contínua. Aqui estão alguns exemplos de variáveis aleatórias discretas:
- Número de filhos em uma família: Esta variável aleatória conta o número de filhos em cada família. Os valores possíveis são números inteiros não negativos, como 0, 1, 2, 3, e assim por diante.
- Número de caras ao lançar uma moeda várias vezes: Se você lançar uma moeda várias vezes, a variável aleatória aqui é o número de vezes que ela mostra "cara". Os valores possíveis são 0 (nenhuma cara), 1 (uma cara), 2 (duas caras) e assim por diante.
- Número de alunos que passam em um exame: Suponha que há um exame com uma taxa de aprovação de 70%. A variável aleatória aqui é o número de alunos que passam no exame. Os valores possíveis são 0 (nenhum aluno passa), 1 (um aluno passa), 2 (dois alunos passam) e assim por diante.
- Número de acidentes em um cruzamento em um determinado dia: Esta variável aleatória representa o número de acidentes ocorridos em um cruzamento durante um dia. Os valores possíveis são números inteiros não negativos

## 1.1 Valor médio de uma variável aleatória discreta
- O valor esperado, também conhecido como média ponderada ou expectativa matemática, ou ainda esperança matemática é um conceito fundamental na teoria da probabilidade e na estatística. Ele representa uma medida de centro de uma distribuição de probabilidade e fornece uma estimativa do valor médio de uma variável aleatória ao longo de um grande número de repetições de um experimento aleatório.
- Matematicamente, o valor esperado de uma variável aleatória X, denotado como E[X] ou μ, é calculado como a soma do produto dos possíveis valores que a variável aleatória pode assumir e suas respectivas probabilidades de ocorrência. Para uma variável aleatória discreta X, o valor esperado é dado por:

- Xi: são os possíveis valores que a variável aleatória X pode assumir.
- P(X=xi): é a probabilidade de que a variável aleatória X assuma o valor xi.

**Por exemplo, se estivermos jogando um dado justo de seis faces, o valor esperado do resultado do lançamento é calculado como:**
- E[X] =  (1 X 1/6) + (2 X 1/6) + (3 X 1/6) + (4 X 1/6) + (5 X 1/6) + (6 X 1/6) = 3.5

- Isso significa que, em média, esperaríamos obter um resultado de aproximadamente 3.5 após um grande número de lançamentos do dado justo.
- Portanto, o valor esperado representa o valor médio que esperaríamos observar ao longo de múltiplas repetições do experimento aleatório. Em outras palavras, ele representa o valor médio que esperamos para a variável aleatória, considerando todas as possíveis ocorrências e suas respectivas probabilidades.

## 1.2 Variância de uma variável aleatória discreta
- A variância de uma variável aleatória discreta é uma medida de dispersão que indica o quão longe os valores da variável estão distribuídos em relação à média. A  a fórmula da variância Var(X) de uma variável aleatória discreta X pode ser expressa como:

- xi: são os valores possíveis da variável aleatória
- E(X): é o valor esperado (média) da variável aleatória X
- pi: são as probabilidades associadas a cada valor xi
- n: é o número total de valores possíveis da variável aleatória X

**Exemplo:  Um empresário tem um lucro de 100 reais por peça feita sem defeito e de 30 reais por peça feita com defeito. Existe uma probabilidade de 20% de que a peça saia com defeito. Qual lucro esperado? E qual a variância do valor esperado?**
- E(X)=(100×0.8)+(30×0.2)=80+6= 86 reais de lucro esperado
- Variância:
    - Para x1 = 100(peça sem defeito):
        - (x1−E(X))².p1= (100−86)²× 0.8 = 196 × 0.8 = 156.8 
    - Para x2 = 30(peça com defeito):
        - (x2−E(X))².p2= (30−86)²× 0.2 = 3136 x 0.2 = 627.2
- Var(X)=156.8 + 627.2= 784 reais

## 1.3 Função de Distribuição Acumulada (FDA)
- **A função de distribuição acumulada (FDA) para variáveis aleatórias discretas é uma função que fornece a probabilidade de que a variável aleatória seja menor ou igual a um determinado valor**. Para uma variável aleatória discreta X, a função de distribuição acumulada é definida como:

- Ou seja,F(x) é a probabilidade de que a variável aleatória X seja menor ou igual a x.
- **Acumulação de probabilidades: A função de distribuição de probabilidade acumula as probabilidades à medida que avança pelos valores da variável aleatória. Por exemplo, se você quer saber a probabilidade de uma variável aleatória  X ser menor ou igual a 5, a função de distribuição de probabilidade soma todas as probabilidades dos valores de X que são menores ou iguais a 5.**
- **Exemplo**:Estamos interessados em encontrar a probabilidade de obter um número menor ou igual a 2 (ou seja, 1 ou 2) se estivermos jogando um dado justo de seis faces:
Para x=2:
F(2)=P(X≤2)
- Há duas faces que mostram 1 ou 2. Portanto, a probabilidade é 2/6, 1/3;
- Então, F(2)=1/3.
- Veja que a distribuição foi acumulada para F(x) = P (X≤2) 1/6 + 1/6 = 1/3. Acumulamos a probabilidade de tirar o 1 ou o 2.

## 1.4 Função de Massa de Probabilidade (PMF) 
- A Função de Massa de Probabilidade (PMF) é uma função que descreve a distribuição de probabilidade de uma variável aleatória discreta. Em outras palavras, **ela nos diz a probabilidade de que uma variável aleatória assuma um determinado valor específico.**
- Variáveis Aleatórias Discretas: A PMF é usada somente para variáveis aleatórias discretas.
- Para cada valor possível da variável aleatória discreta, a PMF atribui a probabilidade de que a variável aleatória assuma esse valor. Assim, a PMF é frequentemente expressa como uma função que mapeia cada valor possível da variável aleatória para sua probabilidade correspondente.

## 1.5 Distribuição Uniforme Discreta
- A distribuição uniforme discreta é um dos modelos mais simples de distribuição de probabilidade para variáveis aleatórias discretas. Ela é frequentemente usada quando todas as categorias ou valores possíveis têm a mesma probabilidade de ocorrência.
- Por exemplo, imagine um dado justo de seis lados. Se você lançá-lo, cada número de 1 a 6 tem a mesma probabilidade de sair, ou seja, 1/6. Isso é um exemplo de distribuição uniforme discreta, pois cada face do dado tem a mesma probabilidade de ocorrer, e esses valores são distintos e contáveis. A fórmula geral para calcular a probabilidade de um evento em uma distribuição uniforme discreta é simplesmente dividir 1 pelo número total de eventos possíveis.


**Exemplo:** Uma rifa tem 100 bilhetes numerados de 1 a 100. Tenho 5 bilhetes consecutivos numerados de 21 a 25, e meu colega tem outros 5 bilhetes, com os números 1, 11, 29, 68 e 93. Quem tem maior probabilidade de ser sorteado?  
**Resposta:** Muitas pessoas podem pensar que espalhar o número é melhor e aumenta a chance de ganhar o prêmio. Entretanto, se estamos falando de uma rifa honesta, temos para cada bilhete a chance 1/100 de ganhar o prêmio. Como eu e meu colega temos 5 bilhetes, ambos temos prbabilidade de 5/100 de ganhar o prêmio.

## 1.6 Distribuição de Bernoulli
- A distribuição de Bernoulli é uma distribuição de probabilidade discreta que **modela um experimento com apenas dois resultados possíveis: sucesso (geralmente representado por 1) ou fracasso (geralmente representado por 0).**
- Para um único ensaio ou experimento, a distribuição de Bernoulli é caracterizada por um parâmetro p, que representa a probabilidade de sucesso (ou seja, a probabilidade de obter o resultado 1) e 1−p, que é a probabilidade de fracasso (ou seja, a probabilidade de obter o resultado 0).
- A função de probabilidade de Bernoulli é dada por:



**Onde:**
- P(X=x)é a probabilidade de obter o resultado x,
- P é a probabilidade de sucesso (resultado 1),
- 1−p é a probabilidade de fracasso (resultado 0).

## 1.7 Distribuição Binomial
- A distribuição binomial é uma distribuição de probabilidade discreta que descreve o número de sucessos em uma sequência fixa de tentativas independentes, onde cada tentativa tem apenas dois resultados possíveis: sucesso ou fracasso. Ela é chamada de **binomial** porque envolve duas possibilidades, sucesso ou fracasso.

- Os principais parâmetros da distribuição binomial são:
    - n: o número total de tentativas ou ensaios.
    - p: a probabilidade de sucesso em cada tentativa.
- A função de probabilidade da distribuição binomial é dada por:



- P(X=k)é a probabilidade de obter exatamente k sucessos em n tentativas,
- (n/k)é o coeficiente binomial, que representa o número de maneiras de escolher k sucessos em n tentativas e é calculado como:
    - n!/k!(n-k)!
- P é a probabilidade de sucesso em uma tentativa individual,
- (1−p) é a probabilidade de fracasso em uma tentativa individual.
- k é o número de sucessos desejados.

A distribuição binomial é usada para modelar situações em que há apenas dois resultados possíveis em cada tentativa, como lançamentos de moedas, ensaios clínicos, teste de produtos, entre outros.  

**Exemplo:** Imagine um evento aleatório dado pelo lançamento de uma moeda honesta e homogênea. Se esta moeda for lançada 4 vezes, qual é a probabilidade de sair pelo menos 3 caras?

3 caras:
- p = Sair cara 1/2, probabilidade de sucesso em uma tentativa individual
- k = 3, número de sucessos desejados
- P(x=3) = (4/3) x 0.5³ x (1-0.5)^4-3
- P(x=3) = (4/3) x 0.125 x 0.5¹
- P(x=3) = (4/3) x 0.125 x 0.5
- P(x=3) = (4/3) x 0.0625

Fatorial: 4!/(3! x (4-3)! =  4 x 3! /(3! x 1) -> Cortamos os dois 3!  
Fatorial: 4/1 = 4   
Resultado: 4  x 0.0625= 0.25

4 caras:
- p = Sair cara 1/2, probabilidade de sucesso em uma tentativa individual
- k = 4, número de sucessos desejados
- P(x=3) = (4/4) x 0.5^4 x (1-0.5)^4-4
- P(x=3) = (4/4) x 0.0625 x 0.5^0
- P(x=3) = (4/4) x 0.0625 x 1
- P(x=3) = (4/4) x 0.0625

Fatorial: 4!/(4! x (4-3)! =  4!/(4! x 1) -> Cortamos os dois 4!  
Fatorial= 1  
Resultado: 1  x 0.0625= 0.0625

**Somando as probabilidades =0.25+ 0.0625=0.3125**


<span style="color: black; background-color: lightgreen;">**Exemplo da aplicação da Função de Massa de Probabilidade (PMF) para distribuição Binomial, utilizando a função dbinom:**</span>
- A probabilidade de um paciente com um ataque cardíaco morrer do ataque é de 0.04, (ou seja, 4 de 100 morrem do ataque).Suponha que tenhamos 5 pacientes que sofrem um ataque cardíaco, qual é a probabilidade de que todos sobrevivam?
- Para este exemplo, vamos chamar um sucesso um ataque fatal (p = 0.04). 
- Temos n = 5 pacientes e queremos saber a probabilidade de que todos sobrevivam ou, em outras palavras, que nenhum seja fatal (0 sucessos).
    - k = 0, número de sucessos que estamos interessados, ou seja, nenhum paciente morrer do ataque. 
    - p = 0.04
    - n = 5
    - **Utilizaremos a função dbinon** pois estamos interessados em calcular a probabilidade de que todos os 5 pacientes sobrevivam.
    
- A função **dbinom** em R é uma função específica da linguagem de programação R usada para calcular a probabilidade de uma variável aleatória discreta seguir uma distribuição binomial. **Ela permite calcular a probabilidade de obter um número específico de sucessos em um determinado número de tentativas, ou seja, ela calcula a função de massa de probabilidade (PMF) para a distribuição binomial**
- dbinom(x, size, prob):
    - x: O número de sucessos desejado, nosso K.
    - size: O número total de tentativas.
    - prob: A probabilidade de sucesso em uma única tentativa.

```r
# Prababilidade de um paciente com pcr vir á óbito.
dbinom(0, 5, 0.04)
```

**Portanto, a probabilidade de que todos os 5 pacientes sobrevivam após um ataque cardíaco é aproximadamente 0.8154, ou seja, cerca de 81.54%.**

### Gráfico das probabilidades da distribuição binomial no R

```r
# Definindo os parâmetros da distribuição binomial
n <- 5  # número total de pacientes

# Calculando as probabilidades para diferentes números de sucessos
k_valores <- 0:n  # de 0 ao n 
prob <- dbinom(k_valores, size = n, prob = 0.04)

# Criando o gráfico de barras
barplot(prob, names.arg = k_valores, col = '#7B68EE', 
        main = 'Probabilidade de Sobrevivência após Ataque Cardíaco',
        xlab = 'Número de Sobreviventes',
        ylab = 'Probabilidade')
```

<span style="color: black; background-color: lightgreen;">**Exemplo da aplicação da Função de Distribuição Acumulada (FDA) para distribuição de Binomial, utilizando a função pbinom:**</span>


**A função pbinom() é uma função em R que calcula a probabilidade acumulada de uma distribuição binomial. Ela é útil quando você está interessado na probabilidade de obter até um certo número de sucessos em um conjunto de tentativas independentes. Estamos interessados em calcular  a Função de Distribuição Acumulada (FDA).** 
**Exemplo:** Suponha que haja 12 questões de múltipla escolha em um questionário de inglês. Cada questão tem cinco respostas possíveis e apenas uma delas está correta. Encontre a probabilidade de ter quatro ou menos respostas corretas se um aluno tentar responder a cada pergunta aleatoriamente.

```r
# Como apenas uma das cinco respostas possíveis está correta, a probabilidade de responder corretamente é 1/5 = 0.2.
# Calcular a probabilidade de ter quatro ou menos respostas corretas
# numero de sucessos: 12, numero de tentativas: 12, probabilidade 0.2
pbinom(4, size=12, prob=0.2)
```

**Portanto, a probabilidade de ter quatro ou menos respostas corretas se um aluno responder a cada pergunta aleatoriamente é aproximadamente 0.93.**
**Calculando novamente a probabilidade de sair pelo menos 3 caras ao lançar uma moeda 4 vezes**
- Como a função pbinom() calcula a probabilidade acumulada de obter **até** um certo número de sucessos em um conjunto de tentativas independentes, passaremos como argumento k, o valor 2, neste caso, 2 ou menos caras, pois estamos interessados em no mínimo 3 caras.
- Se passarmos diretamente '3 caras' como o número de sucessos desejados, a função pbinom() calculará a probabilidade acumulada de obter até 3 caras, não necessariamente no mínimo 3 caras.
    - Por exemplo, ao passar 3 como o número de sucessos desejados para pbinom(), ela retornará a probabilidade acumulada de obter 0, 1, 2 ou 3 caras em 4 lançamentos. Isso não é exatamente o que queremos quando buscamos a probabilidade de obter no mínimo 3 caras.
- Em também setaremos o argumento **lower.tail:** Um argumento booleano que indica se a função de distribuição deve ser avaliada na cauda inferior (TRUE) ou na cauda superior (FALSE) da distribuição. Neste caso, estamos interessados na probabilidade de obter no mínimo 3 caras, então definimos lower.tail como FALSE.


```r
# Probabilidade de obter pelo menos 3 caras
pbinom(2, size = 4, prob = 0.5, lower.tail = FALSE)
```

## 1.8 Distribuição de Poisson

- A distribuição de Poisson é uma distribuição de probabilidade discreta que descreve o número de eventos que ocorrem em um intervalo fixo de tempo ou espaço, assumindo que esses eventos ocorram de forma independente e com uma taxa média conhecida.
- A distribuição de Poisson é caracterizada por um único parâmetro, denotado por λ (lambda), que representa a taxa média de ocorrência do evento por unidade de tempo ou espaço. A função de probabilidade de uma variável aleatória X que segue uma distribuição de Poisson é dada por:


**Onde:**
- P(X=k):é a probabilidade de que o número de eventos seja exatamente k.
- e: é a base do logaritmo natural (aproximadamente 2.71828), número de euler.
- K: é o número de ocorrências do evento.
- λ: é a taxa média de ocorrência do evento.
- k!: é o fatorial de k, número de eventos que ocorrem.

**Exemplo:** Um telefone recebe, em média, cinco chamadas por minuto. Supondo que a distribuição de Poisson seja adequada nessa situação, qual a probabilidade de que o telefone não recebe chamadas durante o intervalo de um minuto?

- Solução:
    - P(X=0) é a probabilidade de não ocorrer nenhum evento = 0
    - e é a base do logaritmo natural (aproximadamente 2.71828)
    - λ é a taxa média de ocorrência de eventos no intervalo de interesse = 5
- P(X=0) = e^-5 x 5^0/ 0!
- P(X=0) = e^-5 x 1/1 -> cortamos os dois 1;
- P(X=0) = e^-5
- P(X=0)≈0.00674

**Portanto, a probabilidade de que o telefone não receba chamadas durante o intervalo de um minuto é aproximadamente 0.0067, ou seja, cerca de 0.67%.**  

<span style="color: black; background-color: lightgreen;">**Exemplo da aplicação da Função de Massa de Probabilidade (PMF) para distribuição de Poison, utilizando a função dpois:**</span>:  
**Considere um processo que têm uma taxa de 0,5 defeitos por unidade. Qual a probabilidade de um processo apresentar duas unidades defeituosas? E apresentar nenhum defeito?**

- dpois(x, lambda);
- x: O valor para o qual você deseja calcular a probabilidade;
- lambda: O parâmetro λ (lambda), que representa a taxa de eventos por unidade;

```r
# Probabilidade de um processo apresentar defeitos em duas unidades, sendo que a taxa é 0,5 defeitos por unidade em um processo.
dpois(2, 0.5)
```

```r
# Probabilidade de um processo apresentar nenhum defeito, sendo que a taxa é 0,5 defeitos por unidade em um processo.
dpois(0, 0.5)
```

<span style="color: black; background-color: lightgreen;">**Exemplo da aplicação da Função de Distribuição Acumulada (FDA) para distribuição de Poison, utilizando a função ppois:**</span>:
**Se houver 12 carros atravessando uma ponte por minuto, em média, encontre a probabilidade de ter 15 ou mais carros cruzando a ponte em um determinado minuto. E a probabilidade de ter 14 ou menos carros atravessando a ponte em um determinado minuto**

- q: valor de interesse:
- lambda: taxa média de eventos por tempo;
- lower.tail: Este argumento indica que queremos calcular a probabilidade da 'cauda superior' ou 'cauda inferior' da distribuição de Poisson, ou seja, a probabilidade de observar valores maiores ou iguais ao valor de interesse, ou menores ou iguais ao valor de interesse.

```r
#Quinze carros ou mais
ppois(15, lambda=12, lower.tail = FALSE)  # upper tail 
```

```r
#14 ou menos carros
ppois(14, lambda=12)   # lower tail 
```
